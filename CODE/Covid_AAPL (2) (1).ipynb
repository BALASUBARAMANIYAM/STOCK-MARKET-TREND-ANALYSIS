```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tcn import TCN
import pandas_ta as ta

# Load data
df = pd.read_excel("AAPL_stock_data.xlsx", skiprows=1)
df.columns = ["Date", "Open", "High", "Low", "Close", "Volume"]
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df.dropna(subset=["Date"], inplace=True)
df.set_index("Date", inplace=True)

# Convert to numeric
for col in ["Open", "High", "Low", "Close", "Volume"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")
df.dropna(inplace=True)

# Add indicators
df["RSI"] = ta.rsi(df["Close"], length=14)
df.ta.macd(close="Close", fast=12, slow=26, signal=9, append=True)
df["MACD"] = df["MACD_12_26_9"]
df["MACD_Signal"] = df["MACDs_12_26_9"]
bb = ta.bbands(df["Close"], length=20)
df["BB_Upper"], df["BB_Lower"] = bb["BBU_20_2.0"], bb["BBL_20_2.0"]
df["ATR"] = ta.atr(df["High"], df["Low"], df["Close"], length=14)
df.dropna(inplace=True)

# Scale
features = ["Close", "Open", "High", "Low", "Volume", "RSI", "MACD", "MACD_Signal", "BB_Upper", "BB_Lower", "ATR"]
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[features])

# Sequence creator
def create_sequences(data, seq_len=60):
    X, y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i:i+seq_len])
        y.append(data[i+seq_len, 0])
    return np.array(X), np.array(y)

seq_len = 60
train_cutoff_date = pd.to_datetime("2020-09-15")
train_end = df.index.get_loc(train_cutoff_date)
X_train, y_train = create_sequences(scaled[:train_end], seq_len)
X_test, y_test = create_sequences(scaled[train_end:], seq_len)

# TCN model
tcn_model = Sequential([
    TCN(input_shape=(seq_len, X_train.shape[2])),
    Dropout(0.2),
    Dense(50, activation="relu"),
    Dense(1)
])
tcn_model.compile(optimizer="adam", loss="mean_squared_error")
tcn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict TCN output for training
tcn_train_out = tcn_model.predict(X_train).reshape(-1, 1, 1)
tcn_test_out = tcn_model.predict(X_test).reshape(-1, 1, 1)

# LSTM model
lstm_model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(1, 1)),
    Dropout(0.2),
    LSTM(100),
    Dropout(0.2),
    Dense(50, activation="relu"),
    Dense(1)
])
lstm_model.compile(optimizer="adam", loss="mean_squared_error")
lstm_model.fit(tcn_train_out, y_train, epochs=50, batch_size=32, validation_data=(tcn_test_out, y_test))

# Predict business days from 2020-09-15 to 2020-11-15
start_date = pd.to_datetime("2020-09-15")
end_date = pd.to_datetime("2020-11-15")
future_dates = pd.date_range(start=start_date, end=end_date, freq='B')
future_days = len(future_dates)

# Predict forward
future_preds = []
last_seq = scaled[train_end - seq_len:train_end]

for _ in range(future_days):
    tcn_input = last_seq.reshape(1, seq_len, X_train.shape[2])
    tcn_output = tcn_model.predict(tcn_input).reshape(1, 1, 1)
    lstm_pred = lstm_model.predict(tcn_output)[0][0]
    future_preds.append(lstm_pred)
    next_seq = last_seq.copy()
    next_seq = np.roll(next_seq, -1, axis=0)
    next_seq[-1, 0] = lstm_pred
    last_seq = next_seq

# Inverse scale predictions
padded = np.zeros((future_days, len(features)))
padded[:, 0] = future_preds
predicted_close = scaler.inverse_transform(padded)[:, 0]

# Download actual prices
actual_df = yf.download("AAPL", start=start_date, end=end_date)
actual_df = actual_df.reindex(future_dates)
actual_close = actual_df["Close"].values

# Replace predicted with actual when available
final_output = predicted_close.copy()
for i in range(future_days):
    if not np.isnan(actual_close[i]):
        final_output[i] = actual_close[i]

# Plot results
plt.figure(figsize=(14, 6))
plt.plot(future_dates, predicted_close, label="Predicted", linestyle="--", color="green", alpha=0.7)
plt.plot(future_dates, actual_close, label="Actual", color="blue")
plt.plot(future_dates, final_output, label="Final (Updated)", color="orange")
plt.title("TCN + LSTM AAPL Prediction (2020-09-15 to 2020-11-15)")
plt.xlabel("Date")
plt.ylabel("AAPL Close Price")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
